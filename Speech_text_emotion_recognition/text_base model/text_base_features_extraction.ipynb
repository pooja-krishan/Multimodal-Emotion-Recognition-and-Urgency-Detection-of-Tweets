```python
import pandas as pd
import matplotlib.pyplot as plt
from collections import Counter
from sklearn import preprocessing
from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import StandardScaler
from nltk.tokenize import sent_tokenize,word_tokenize
from nltk import pos_tag
from nltk.corpus import stopwords
import numpy as np
import seaborn as sns
import nltk as nl
import spacy
import re
from textblob import Word
nl.download('punkt')
nl.download('averaged_perceptron_tagger')
nlp = spacy.load("en_core_web_sm")
nl.download('stopwords')
```

```python
text_df = pd.read_csv('IEMOCAP_TEXT/preprocessed_transcription_text.csv')
```

```python
text_df.head()
```

```python
def awl(text):
  word = word_tokenize(text)
  return (sum(len(words) for words in word)/len(word))

def asl(text):
  sentences = sent_tokenize(text)
  words = word_tokenize(text)
  average_sentence_length = len(words) / len(sentences)
  return average_sentence_length

def now(text):
  return len(word_tokenize(text))

def nvb(text):
  x = word_tokenize(text)
  sum = 0
  st = 'VB'
  tokens_tag = pos_tag(x)
  counts = Counter(tag for word,tag in tokens_tag)
  for i in counts.keys():
    if st in i:
      sum += counts[i]
  return sum

def naj(text):
  x = word_tokenize(text)
  sum = 0
  st = 'JJ'
  tokens_tag = pos_tag(x)
  counts = Counter(tag for word,tag in tokens_tag)
  for i in counts.keys():
    if st in i:
      sum += counts[i]
  return sum

def npv(text):
  x = nlp(text)
  passive_toks=[tok for tok in x if (tok.dep_ == "nsubjpass") ]
  if passive_toks != []:
    return len(passive_toks)
  return 0

def nst(text):
  return len(sent_tokenize(text))

def ntp(text):
  words = text.split()
  words = [word.lower() for word in words]  #convert to lower case
  words = [re.sub(r'[^A-Za-z0-9]+', '', word) for word in words]  # remove puntuation signs
  count = 0
  for word in words:
    each_word = Word(word)
    result = each_word.spellcheck()
    if each_word != result[0][0]:
      count += 1
  return count

def tpr(text):
  num_typo = ntp(text)
  num_word = now(text)
  return num_typo/num_word

def cdv(text):
  num_word = now(text)
  word = word_tokenize(text)
  word = [w.lower() for w in word]  #convert to lower case
  word = [re.sub(r'[^A-Za-z0-9]+', '', w) for w in word]  # remove puntuation signs
  word[:] = [x for x in word if x]      # remove empty string
  word_without_sw = [w for w in word if not w in stopwords.words()]
  if(len(word_without_sw) == 0):
    return 0
  else:
    return num_word / len(word_without_sw)
```

```python
text_df['AWL'] = text_df['processed_text'].apply(lambda x: awl(x))
```

```python
text_df['ASL'] = text_df['processed_text'].apply(lambda x: asl(x))
text_df['NOW'] = text_df['processed_text'].apply(lambda x: now(x))
text_df['NVB'] = text_df['processed_text'].apply(lambda x: nvb(x))
text_df['NAJ'] = text_df['processed_text'].apply(lambda x: naj(x))
text_df['NPV'] = text_df['processed_text'].apply(lambda x: npv(x))
text_df['NST'] = text_df['processed_text'].apply(lambda x: nst(x))
```
